## Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:  
1. country  
2. year  
3. diagnosis method: mapping should be: 

```
  {'newsp':'Pulmonary Smear (+)'
  ,'newsn':'Pulmonary Smear (-)'
  ,'newep':'Extrapulmonary'
  ,'newrel':'Relapse'}   
 ```

4. gender: male or female.   
5. age: lower_age - higher_age, e.g., 0-14  
6. number of cases  

__Note__: You can use `replace` (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)

````python
df.tail()
````

````python
# conver the pivot table back to long form from wide form
# the remaining group type columns will be melted with respect to country and year, and their values will be in the number_of_cases column
melted_df = df.melt(id_vars=['country', 'year'], var_name='group_type', value_name='number of cases')
melted_df
````

````python
# we don't need null values now
melted_df = melted_df.dropna()
melted_df = melted_df.reset_index(drop=True)
````

````python
melted_df.group_type.values[0]
````

<b> if we look here. 
    new_sp -> newsp (meaning pulmonary smear (+))
    similarly, m means male
    014 means age_group 0-14
</b>

I will use regular expressions to extract those values

````python
import re

# this is our pattern

'''
here () means one group. since we are extracting diagnosis method, gender and age groups, we have 3 groups. lets look at them
    1. (new_?[a-z]+) -> starts with new, followed by _ (? means there can be _ or not. its optional). followed 
                - by [a-z]+ meaning a-z smaller cap characters at least once. this gives us our diagnosis.
                - followed by a compulsory _
    2. ([mf]) -> another group where one only one character m or f is selected. m being male and f being female.
    3. (\d+) -> final group where there can be more than 1 numeric values. same as [0-9]+ for the age group

'''
pattern = r"(new_?[a-z]+)_([mf])(\d+)"

match = re.search(pattern, 'new_sp_m014')

if match:   
    print("diagnosis method: ",  match.group(1) )
    print("gender: ", match.group(2) )
    print("age grroup:", match.group(3))
````

We can do this in pandas using .str.extract() method directly

````python
# accessing the new features
melted_df[["diagnosis method", "gender", "age"]] = melted_df["group_type"].str.extract(r"(new_?[a-z]+)_([mf])(\d+)")
````

````python
melted_df.head()
````

````python
melted_df["diagnosis method"].unique()
````

````python
# lets map the diagnosis method to original values
diagnosis_map = {
    'new_sp': 'Pulmonary Smear (+)',
    'new_sn': 'Pulmonary Smear (-)',
    'new_ep': 'Extrapulmonary',
    'newrel': 'Relapse'
}

# gender mapping
gender_map = {
    "m": "Male",
    "f": "Female"
}

melted_df["diagnosis method"] = melted_df["diagnosis method"].map(diagnosis_map)
melted_df["gender"] = melted_df["gender"].map(gender_map)
````

````python
# lets convert ages to age brackets using slicing
melted_df["age"] = melted_df["age"].apply(lambda x : x[0:-2]+ "-" + x[-2:] if x != '65' else "65+")
````

````python
# rearranging columns
melted_df = melted_df[["country", "year", "diagnosis method", "gender", "age", "number of cases"]]
````

## Question 3: Create a graph that shows when countries started to report TB cases.

````python
# lets check when each country started reporting TB
# the reason we are filtering based on number of cases is because we want to get year and name of countries -
# - sorted in ascending order based on year to get when countries started reporting TB cases.
countries_and_their_first_reporting_year = melted_df[melted_df["number of cases"] > 0].groupby("country")["year"].min().reset_index().sort_values("year", ascending=True)
countries_and_their_first_reporting_year
````

````python
plt.figure(figsize=(15, 7))
plt.scatter(countries_and_their_first_reporting_year['year'], countries_and_their_first_reporting_year['country'])

plt.xlabel('Year when a country started reporting TB', fontsize=10)
plt.ylabel('Country Name', fontsize=12)
plt.title(' TB Case Reporting by Country with respect to year', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

Well. Very few cuntries actually had TB reports in the early 1980s. Howeer, since mid 1990s, there is an exponential increase in TB reports where majority of countries reported TB cases. Lets make a histogram as this scatterplot is very congested as lots of countires have reported TB cases and its not possible to keep every single country name in the axes.

````python
# creaitng a histogram
plt.figure(figsize=(12, 6))
sns.histplot(countries_and_their_first_reporting_year['year'], bins=20, kde=True, discrete=True)
plt.ylabel("Number of Countries")
plt.xlabel("Year")
plt.title("Histogram that represents the number of Countries reporting TB cases")
plt.grid(True)
plt.show()
````

if we look at above histogram, it shows that the earliest reporting was in 1980s with very few reports. However, in the mid 1990s, there is a huge surge of TB reports. What's interesting is that in 1995, more than 120 countries reported TB cases. After that, the TB reports have been pretty low, similar to before mid 1990s.

<b> Therefore, using the histogram above, we can see when countries started reportng TB cases. A lot of countries reported TB cases in the mid 1990s, in the year 1995. Perhaps there was a TB epidemic around the world which lead to this.
